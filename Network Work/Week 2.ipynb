```python
# Exercise 3: Conditions

# Checking to see if a number is positive, negative, or zero
num = float(input("Enter a number: "))
if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")

# Checking to see if a word starts with a vowel or a consonant
word = input("Enter a word: ")
if word[0].lower() in 'aeiou':
    print("The word starts with a vowel.")
else:
    print("The word starts with a consonant.")

# Checking to see if two numbers are equal
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
if num1 == num2:
    print("Both numbers are equal.")
else:
    print("The numbers are not equal.")
```

```python
# Exercise 4: Loops

# Loop printing numbers from 1 to 10
print("Numbers from 1 to 10:")
for i in range(1, 11):
    print(i, end=' ')

print("\n")

# Loop printing even numbers from 2 to 10
print("Even numbers from 2 to 10:")
for i in range(2, 11, 2):
    print(i, end=' ')

print("\n")

# While loop: print the first 5 numbers of the Fibonacci sequence
fib_num = [0, 1]
while len(fib_num) < 5:
    fib_num.append(fib_num[-1] + fib_num[-2])
print("Here are the first 5 Fibonacci numbers: ", fib_num)
```

```python
# Exercise 5: Functions

# Function to return the sum of two numbers
def add_numbers(x, y):
    return x + y

print("Sum of 3 and 5 is:", add_numbers(3, 5))

# Function to return the largest number in a list
def largest_number(lst):
    return max(lst)

print("Largest number in [3, 7, 2, 9, 5]:", largest_number([3, 7, 2, 9, 5]))

# Function to count how many vowels in a string
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = sum(1 for char in s if char in vowels)
    return count

print("Number of vowels in 'Hello World':", count_vowels("Hello World"))
```

```python
# Exercise 6: Password Strength Checker

def check_password_strength(password):
    special_characters = "!@#$%^&*()-+?_=,<>/"
    messages = []

    if len(password) < 8:
        messages.append("Password must be at least 8 characters long.")
    if not any(char.isupper() for char in password):
        messages.append("Password must contain at least one uppercase letter.")
    if not any(char.islower() for char in password):
        messages.append("Password must contain at least one lowercase letter.")
    if not any(char in special_characters for char in password):
        messages.append("Password must contain at least one special character.")

    if messages:
        print("\n".join(messages))
    else:
        print("Password is strong.")

# Testing the password checker
print("Test 1:")
check_password_strength("Pa$$word123")
print("\nTest 2:")
check_password_strength("password")
```

```python
# Exercise 7: Caesar Cipher

def caesar_cipher(message, key, mode="encrypt"):
    result = ""
    for char in message:
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            shift = key if mode == "encrypt" else -key
            result += chr((ord(char) - base + shift) % 26 + base)
        else:
            result += char
    return result

# Testing the created Caesar cipher
encrypted_word = caesar_cipher("HELLO", 3, mode="encrypt")
print("Encrypted message:", encrypted_word)
decrypted_word = caesar_cipher(encrypted_word, 3, mode="decrypt")
print("Decrypted message:", decrypted_word)
```

```python
# Exercise 8: Hangman Game

import random

def hangman():
    word_list = ["python", "java", "javascript"]
    word = random.choice(word_list)
    guessed = ["_"] * len(word)
    attempts = 6
    guessed_letters = set()

    while attempts > 0 and "_" in guessed:
        print("\nWord:", " ".join(guessed))
        guess = input("Guess a letter: ").lower()

        if guess in guessed_letters:
            print("You've already guessed that letter.")
            continue

        guessed_letters.add(guess)

        if guess in word:
            for i in range(len(word)):
                if word[i] == guess:
                    guessed[i] = guess
        else:
            attempts -= 1
            print("Incorrect! Attempts left:", attempts)

    if "_" not in guessed:
        print("\nCongratulations, you guessed the word:", word)
    else:
        print("\nGame over! The word was:", word)

hangman()
```
